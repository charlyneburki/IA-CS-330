```python
class AnalyseArbre:
    
     def __init__(self, arbre):
        
        self.attribut = arbre.attribut
        self.enfants = arbre.enfants
        self.regles = []
        self.parcours = []
        self.trace = []
    
     def generer_regle(self, level=0):
        """ Génère les règles à partir de l'arbre.
        """
        regle = []
        if self.terminal():
            #noeud est terminal il faut alors donner le resultat et l'ajouter dans le tracé et commencer à iterer
            #en arrière. 
            #il faut aussi revenir en arrière de 1 dans la liste de parcours et recommencer à parcourir à partir de
            #celui-ci
            print('pretty')
            result = '=> {}\n'.format(self.classe().upper())
            regle.append(result)
            
        else:
            for valeur,enfant in self.enfants.items():
                condition = ' {} = {} '.format(self.attribut, valeur.upper())
                regle.append(condition)
                regle.append(enfant.generer_regle(level+1))
                self.regles.append(regle)
                #regle = []
        
        return regle
    
     def terminal(self):

        return self.enfants is None
    
     def classe(self):
        """ Si le noeud est terminal, retourne la classe des données qui\
            tombent dans la sous-classification (dans ce cas, toutes les\
            données font partie de la même classe. 
        """

        if self.terminal():
            return self.donnees[0][0]
    
    
```
